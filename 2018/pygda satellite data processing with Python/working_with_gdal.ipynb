```python
eo_data = gdal.Open('LC08_L1TP_186024_20160630_20170323_01_T1/LC08_L1TP_186024_20160630_20170323_01_T1_B2.TIF')
```

```python
band_src = eo_data.GetRasterBand(1)
```

```python
gdal.GetDataTypeName(band_src.DataType)
```

```python
band_src.GetMinimum()
```

```python
band_src.GetMaximum()
```

```python
# del eo_data
```

```python
img1 = band_src.ReadAsArray(0, 0, eo_data.RasterXSize, eo_data.RasterYSize)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

ax.imshow(img1, cmap='gray')
```
